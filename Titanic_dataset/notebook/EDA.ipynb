```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic  = pd.read_csv('Cleaned_titanic.csv')
titanic.head()
```

```python
#  total passengers survived -> 342
plt.figure(figsize=(6,4))
(titanic['Survived'].value_counts()).plot(kind='bar',color='black')

plt.title('Survival distribution')
plt.xlabel('0-> not survived 1-> survived')
plt.ylabel('No of Passengers')
plt.show()
```

```python
#  total male and female
#  male -> 577
#  female ->  314
plt.figure(figsize=(6,4))
titanic['Sex'].value_counts().plot(kind='bar',color=['blue','pink'])

plt.ylabel('No of Passenger')
plt.title('Total Male and Female')
plt.show()
```

```python
# passenger class distribution
# 3-> 491
# 1-> 216
# 2-> 184
plt.figure(figsize=(6,4))
titanic['Pclass'].value_counts().sort_index().plot(kind='bar')

plt.ylabel('No of passenger')
plt.title('PASSENGER CLASS DISTRIBUTION')
plt.show()
```

```python
# survival by gender
#  female -> 233
#  male -> 109
plt.figure(figsize=(6,4))
(titanic.groupby('Sex')['Survived'].mean()*100).plot(kind='bar')

plt.title('SURVIVAL BY GENDER PERCENTAGE')
plt.ylabel('PASSENGER PERCENT')
plt.show()
```

```python
#  survival by passenger class
plt.figure(figsize=(6,4))
(titanic.groupby('Pclass')['Survived'].mean()*100).plot(kind='bar')

plt.title('SURVIVAL BY passenger class PERCENTAGE')
plt.ylabel('PASSENGER PERCENT')
plt.show()
```

```python
# Age  distribution
plt.figure(figsize=(8,5))
plt.hist(titanic['Age'],bins=20,color='black')
plt.title('Age distribution')
plt.xlabel('Age')
plt.ylabel('No of Passenger')

plt.show()
```

```python
#  fare distribution
plt.figure(figsize=(8,5))
plt.hist(titanic['Fare'],bins=30,color='black')
plt.title('Fare distribution')
plt.xlabel('Fare')
plt.ylabel('No of Passenger')
plt.show()
```

```python
# Embarked distribution
(titanic['Embarked'].value_counts()).plot(kind='bar',color=['red','green','blue'])
plt.title('Embarked Distribution')

plt.ylabel('No of Passenger')
plt.show()
```

```python
# Key Insights

# 1. More passengers died than survived.
# 2. Female passengers had a much higher survival rate than males.
# 3. First-class passengers had the highest survival rate.
# 4. Most passengers traveled in third class.
# 5. Most passengers were between 20 and 40 years old.
# 6. Fare distribution was right-skewed, with a few passengers paying very high fares.
```

```python
# Conclusion

# The analysis suggests that gender and passenger class had a significant impact on survival. Female passengers and first-class passengers were more likely to survive than male and third-class passengers.
```
